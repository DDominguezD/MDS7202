````python
# impieza del dataframe
def clean_data(df):
  """Realiza una limpieza de los datos eliminando duplicados, nulos y qued치ndose
  con notas en el rango [0, 100].

    Parameters
    ----------
    df : pandas.dataframe
        Base de datos a limpiar

    Returns
    -------
    df: 
        pd.dataframe ya limpiado
    """
  # Borrar duplicados
  df.drop_duplicates(subset=['names'], keep='first', inplace=True)
  # Reemplazar '' por nan
  df.replace('', np.nan, inplace=True)
  #Borrar nan
  df.dropna(inplace=True)
  # Notas en rango [0,100]
  df = df[df['math score'].between(0, 100)]
  df = df[df['reading score'].between(0, 100)]
  df = df[df['writing score'].between(0, 100)]
  # Rehacer 칤ndice del dataframe
  df = df.reset_index(drop=True)
  return df
````

### 1.3.1 Transformaci칩n de notas a "Nota Chilena" [0.5 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intenci칩n de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuaci칩n:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuaci칩n entregada, cree una funci칩n que le permita pasar a la escala Chilena las notas que se escapan del rango [1,7] y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
# Nota [0, 100] a [0, 7]
def nota_chilena(dataframe_in, columna):
  """Cambia la nota de 0 a 100 a una escala de 0 a 7 mediante
    la f칩rmula NotaChilena = % de logro * 0.06 + 1 redondeano
    al segundo decimal.

    Parameters
    ----------
    dataframe_in : pandas.dataframe
        Base de datos a realizar el cambio de nota
    columna: str
        Nombre de la columna de dataframe_in que se quiere modificar

    Returns
    -------
    dataframe: 
        pd.dataframe con los cambios realizados a la nota
    """
  dataframe_in[columna] = dataframe_in[columna].apply(lambda x: round(x*0.06+1, 2))
  return dataframe_in
````

### 1.3.2 Obtenci칩n de promedio de alumnos [0.5]

Con todas las notas adjuntadas, la prestigiosa Universidad De la Cacha침a necesita obtener el promedio general de cada uno de los alumnos. Para esto se le solicita que calcule el promedio con cada una de las asignaturas y el promedio se anexe como una nueva columna al final del `DataFrame` original. Cabe se침alar que las notas deben ser redondeadas al segundo decimal.

**To-Do:**
- [ ] Crear funci칩n que nos permite obtener el promedio de cada uno de los alumnos y anexar esta informaci칩n en el `DataFrame` original.


**Respuesta:**

````python
# C치lculo del promedio de alumnos
def alumnos_promedio(dataframe_in):
  """Calcula el promedio simple de las notas redondeadas al segundo decimal.

    Parameters
    ----------
    dataframe_in : pandas.dataframe
        Base de datos que se le agrega la columna del promedio de notas.

    Returns
    -------
    dataframe: 
        pd.dataframe con una columna 'GPA' nueva que contiene el promedio
    """
  # Pasar a nota chilena
  dataframe_in = nota_chilena(dataframe_in, 'math score')
  dataframe_in = nota_chilena(dataframe_in, 'reading score')
  dataframe_in = nota_chilena(dataframe_in, 'writing score')
  # Promedio redondeado
  dataframe_in['GPA'] = round((dataframe_in['math score']+dataframe_in['reading score']+dataframe_in['writing score'])/3, 2)
  return dataframe_in
````

### 1.3.3 Reprobados y Destacados [1.5 Punto, 0.5 Cada DataFrame]

Contentos con su desempe침o, la Universidad les solicita que obtengan los alumnos quienes han reprobado el a침o y los que han tenido desempe침o de destacado. Para esto se considerar치 que un alumno ha reprobado el a침o cuando el promedio de notas es inferior a 4.0 en la Escala Chilena, y se considerar치 destacado si posee notas superiores a 5.8 en todas las asignaturas (o sea notas superiores a 5.8 en `math score`, `reading score` y `writing score`).

Cree una funci칩n que permita obtener tres Dataframes: 
- El primero deber치 poseer solamente los alumnos que han reprobado en la prestigiosa universidad.
- El segundo debe entregar un DataFrame similar al original, pero marcado con una "R" el promedio de notas de los alumnos reprobados.
- El tercero ser치 un Data Frame con los alumnos destacados ordenados de mayor a menor.

**To-Do:**

- [ ] Obtener DataFrame con los alumnos que reprobaron el a침o.
- [ ] Obtener DataFrame similar al original, pero marcados con una "R" el atributo GPA de los alumnos reprobados.
- [ ] Obtener DataFrame con los alumnos destacados ordenados de mayor a menor.

**Respuesta:**

````python
def reprobados_destacados(dataframe_in):
    """Obtiene 3 dataframes; uno con alumnos reprobados, otro con toda la informaci칩n
    y una 'R' a los reprobados y otro con alumnos destacados (que hayan tenido nota
    superior a 5.8 en todas las asignaturas).

    Parameters
    ----------
    dataframe_in : pandas.dataframe
        Base de datos a calcular reprobados y destacados

    Returns
    -------
    dataframe, dataframe, dataframe: 
        3 pd.dataframes con reprobados, alumnos con una 'R' y destacados en orden descendiente.
    """
    # df reprobados
    df_reprobados = dataframe_in[dataframe_in['GPA'].between(0, 4, inclusive='neither')]
    # reindexar
    df_reprobados=df_reprobados.reset_index(drop=True) 
    # df completo con 'R' a los reprobados
    df_R = dataframe_in.copy()
    df_R['GPA'] = df_R['GPA'].apply(lambda x: 'R' if x<4 else x) 
    # reindexar
    df_R=df_R.reset_index(drop=True) 
    # df destacados
    df_destacados = dataframe_in[dataframe_in['math score'].between(5.8, 7, inclusive='right')]
    df_destacados = df_destacados[df_destacados['reading score'].between(5.8, 7, inclusive='right')]
    df_destacados = df_destacados[df_destacados['writing score'].between(5.8, 7, inclusive='right')]
    # orden descendiente
    df_destacados = df_destacados.sort_values(by=['GPA'], ascending=False)
    # reindexar
    df_destacados=df_destacados.reset_index(drop=True) 
    return df_reprobados, df_R, df_destacados
````
