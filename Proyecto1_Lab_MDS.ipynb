## 4. Preprocesamiento

Para esta parte del trabajo, se genera un  `ColumnTransformer` que permita modificar los datos. Estas modificaciones son `OneHotEncoder`, `StandarScaler` y `OrdinalEncoder`.  

Para ver a cuales columnas aplicarles alguna transformación, se analizarán primero sus tipos y si tiene sentido transformar la variable. Es decir, la columna **"id"** y **"fraud_bool"** carecen de sentido para las transformaciones por lo que se mantendrán como vienen en el dataframe. Por otro lado, las columnas numéricas pasarán por un escaamiento, mientras que las de tipo **'Object'**, se verán si poseen pocos valores únicos (indicando que son características categóricas) para pasarlas a *one hot encoder*.  

Por último, se genera un `Pipeline` para guardar las transformaciones y así generar un nuevo dataframe procesado.

### 4.1 Declarar `ColumnTransformer`

```python
# Verificar datos tipo 'object'
print('Valores únicos de las columnas de tipo object')
print(' ')
column_object=list(df.select_dtypes(['object']).columns)
for col in column_object:
    print(col+':', df[col].unique())
```

```python
# Columnas tipo float
column_float=list(df.select_dtypes(['float64']).columns)
column_float=column_float[1:] #No se considera la columna fraud_bool
```

```python
# Preprocesamiento de datos
preprocessing_transformer = ColumnTransformer(transformers=[
    ('OneHotEncoder', OneHotEncoder(sparse_output=False), column_object),
    ('scaler', StandardScaler(), column_float),
    ('ordinal_encoder', OrdinalEncoder(), ['segmentacion_etaria'])], 
    remainder='passthrough')
```

```python
# Crear pipeline
pipeline = Pipeline([('Preprocessing', preprocessing_transformer)])
```

### 4.2 Transformar datos

```python
# Retornar datframe procesado
pipeline.set_output(transform="pandas")
df_preprocesado = pipeline.fit_transform(df)
```

### 6.4 Calcular ratios y responder

```python
total_fraude = len(df_sample_test[df_sample_test['prediction']=='1.0'])
tp_fraude = len(df_sample_test[(df_sample_test['prediction']=='1.0')&(df_sample_test['fraud_bool']=='1.0')])

P = tp_fraude/total_fraude
print('P =', P)
```

```python
total_real_fraude = len(df_sample_test[df_sample_test['fraud_bool']=='1.0'])

R = tp_fraude/total_real_fraude
print('R =', R)
```

**¿Qué significan los ratios y sus valores?**

Los ratios P y R significan precision y recall respectivamente. 

La precisión se define como la proporción de instancias positivas clasificadas correctamente sobre el total de instancias clasificadas como positivas. 
$$Precisión = Verdaderos \ positivos / (Verdaderos \ positivos + Falsos \ positivos)$$

El recall mide la proporción de instancias positivas clasificadas correctamente sobre el total de instancias que realmente son positivas.

$$Recall = Verdaderos \ positivos / (Verdaderos \ positivos + Falsos \ negativos)$$

Estos valores van entre $0$ y $1$, y entre más alto sea su valor, mejor es el resultado de la predicción.
